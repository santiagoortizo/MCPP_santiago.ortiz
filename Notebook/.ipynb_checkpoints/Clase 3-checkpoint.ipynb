En las listas, el orden importa y se mantiene. Es una estructura de datos ordenada.
En Python, las estructuras de datos tienen indexación en base 0. La posición 0 tiene el primer elemento y así ...

```python
lenguajes[-1] # Podemos ver esto de derecha a izquierda, empieza en -1.
```

### Slicing

La sintaxis va desde la posición i hasta el j-1.

```python
# Muestre los primeros tres elementos, no todo.
lenguajes[0:3]
```

```python
lenguajes[0:4]
```

```python
# Se puede especificar donde empieza o donde termina.
lenguajes[:4] #Vaya a la posición 4-1 y de ahí para atras coja todo
```

```python
lenguajes[1:]
```

```python
lenguajes[0:4:2] # Desde la posición 0 hasta la posición 4-1 de dos en dos.
```

```python
lenguajes[0:4:10]
```

```python
# Construir una lista de 10000 números aleatorios entre 0 y 9
import random # Importar un módulo (librería)

N = 10000
random_numbers = [] # Inicializar lista vacía.
for i in range(N): # Para i en [0,9999]. (10000 veces)
    random_numbers.append(random.randint(0, 9)) #El append es para pegar un elemento nuevo a la lista 
    

```

¿Cómo hallar la frecuencia de cada uno de los números?

```python
counter = []
for x in range(0,10):
    frecuencia = random_numbers.count(x) #.count es un método. Sirve para contar las ocurrencias de la variable x
    counter.append(frecuencia) 
```

```python
counter
```

### Ejercicios en clase 3

1. Haga un algoritmo que, utlizando listas, dibuje un rombo sin espacios entre los astericos y que funcione solo con números impares (si el usuario introduce un número par, le indique que debe ser impar y que lo vuelva a intentar).

```python
numero = int(input("Introduzca un número impar: "))
while numero % 2 == 0:
    numero = int(input("Introduzca un número IMPAR: "))
lista = []
espacio = numero
for i in range(numero + 1):
    if i % 2 !=0:
        espacio = espacio - 1
        lista.append(i)
        print((" "*espacio)+("*"*i))
lista.remove(numero)        
for i in lista[::-1]:
    espacio = espacio + 1
    print((" "*espacio)+("*"*i))
    
    
```

2. Escriba un programa que permita crear una lista de palabras y que, a continuación, pida una palabra y diga cuántas veces aparece esa palabra en la lista.

```python
numero = int(input("De cuantas palabras quiere que sea su lista: "))
while numero == 0:
    numero = int(input("Digite un numero diferente a 0: "))
lista_palabras = []
for i in range(numero):
    palabra = input(f"Deme la palabra número {i + 1}: ")
    lista_palabras.append(palabra)
print("Su lista tiene las siguientes palabras: \n", lista_palabras)
palabra2 = input("Deme una palabra: ")
counter2 = []
contador = 0
for x in lista_palabras:
    if x == palabra2:
        contador = contador + 1
if contador == 0:
    print(f"La palabra {palabra2} no está en la lista.")
elif contador == 1:
    print(f"La palabra {palabra2} está solo una vez en la lista.")
else:
    print(f"La palabra {palabra2} está {contador} veces en la lista.")
        
```

3. Escribir un programa que almacene en una lista los siguientes precios, 50, 75, 46, 22, 80, 65, 8, y muestre por pantalla el menor y el mayor de los precios

```python
precios = [50, 75, 46, 22, 80, 65, 8]
mayor, menor = max(precios), min(precios)
print(f"{mayor} es el preciomás alto y {menor} es el precio más bajo")
```

```python
prices = [50, 75, 46, 22, 80, 65, 8]
min1 = max1 = prices[0]
for price in prices:
    if price < min1:
        min1 = price
    elif price > max1:
        max1 = price 
print("El mínimo es " + str(min1)) 
print("El máximo es " + str(max1))
```

```python
numbers = [50, 75, 46, 22, 80, 65, 8]
numbers.sort(reverse = True) 
print("El máximo es: ", numbers[0]) 
print("El mínimo es: ", numbers[-1]) 
```

4. Escribir un programa que pregunte por una muestra de números, separados por comas, los guarde en una lista y muestre por pantalla su media y desviación típica.

```python
numeros = input("Introduzca números separados por una ',': ")
split = numeros.split(",")
print("Está es su lista de números: \n", split)
n = len(split)
suma = 0
sumaq = 0
for i in range(n):
    suma = suma + int(split[i])
    sumq += int(split[i])**2
mean = suma / n
stdev = (sumq/n-mean**2)**(1/2)
print('La media es', mean, ', y la desviación típica es', stdev)
    
```
