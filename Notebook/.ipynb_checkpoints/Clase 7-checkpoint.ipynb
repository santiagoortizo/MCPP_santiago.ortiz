Arreglos de N-dimensión.


```python
import numpy as np # Importar librerias y darle un nombre particular
                   # Es la libreria que permite hacer trabajos númericos en Python
```

```python
# Creamos una lista de lo que se supone sería una matriz
# Cada elemento de la lista es una lista, que se supone que son
# las filas.
m1 = [[0,1,4], [9,16,25], [36,49,64], [81,100,121]] 
```

```python
print(m1) # Note que se parece a una matriz. Pero no es una matriz
          # Sigue siendo una lista cuyos elementos tambien son listas.
```

Suponga que quiere multiplicar por el escalar "2"

```python
print(m1*2) 
# Dado que NO es una matriz. Lo que hara es multiplicarme la lista
# Estoy concatenando la misma lista. Solo me la expande.
```

```python
# Multiplicar por el escalar 2 todos los elementos de la "matriz"
# Método 1:
m2 = []
for i in range(len(m1)):
    l = []
    for j in range(len(m1[0])):
        l.append(m1[i][j]*2)
    m2.append(l)
print(m2)
# Lo que hace este código es multiplicar todos los elementos de la lista
# incluidos los elementos dentro de las listas por 2.
```

```python
# Método 2
# Utilizar list comprehension para reducir el mismo comando
m2 = [[x*2 for x in row] for row in m1]
print(m2)
# Crea una lista m2, despues cada aplica la función a cada x que pertenece
# a row, en donde row es cada uno de los elementos que estan en la
# lista m1. es como el m1[i][j]
```

Ahora utilizando numpy podemos trabajar como si fuera una matriz

```python
# Utilizando np.array
# Recuerde que m1 es una lista cuyos elementos son listas.
ma1 = np.array(m1) # Llamo a numpy con np (así fue como la llame arriba)
                   # Utilizo el método array (arreglo de números)
                   # array([[a,b,c],[d,e,f]]) guarda las filas en formato lista.
                   # Hay qye añadirle [] antes de.
print(ma1) # Note que ma1 se ve más como una matriz
```

```python
type(ma1) # Tipo de ma1
```

```python
ma1*2 # Dado que ma1 ya es una matriz, podemos trabajar normal
```

```python
print(ma1*2)
```

```python
print(np.sin(ma1)) # np.sin() aplica el seno a lo que queramos
```

```python
print(ma1 - 2)
```

```python
print(ma1 ** 2)
```
