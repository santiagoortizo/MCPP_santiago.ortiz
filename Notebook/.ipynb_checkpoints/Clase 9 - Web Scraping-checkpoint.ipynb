#### Explicación
- str(links) para tomar como un string toda lista
- Necesitamos coger todo lo que tenga esa forma de http://...
- Note que es igual para todos, pero después de Paginas empieza a cambiar.
- con el ? hacemos que solo lo haga UNA sola vez. Es decir, va a parar cuando encuentre el ", pero no va a buscar nada más. Ahí ya tiene una observación.

```python
# Sobreescribimos la variable links. Ahora solo serán los links que nos competen.
links = re.findall('"(http://wsp.presidencia.gov.co/Prensa/.+?)"', str(links))
```

A pesar de que esto no es ilegal, muchas páginas no quieren que haya gente chismosa. Por eso hay que relentizar el problema.

```python
import time
def try_request(url):
    html = requests.get(url).text # Meter la URL y extraer el HTML
    time.sleep(1) # Demorese 1 segundo haciendo ese proceso.
    return html
```

Note que la función try_request lo que hará es obtener todo el código html de cada una de las páginas que quiera visitar.

```python
for link in links:
    html = try_request(link) # Guarde el código html de cada url
    print(html[0:1000]) # Imprimame los 1000 elementos que tenga.
                        # recuerde que si tiene menos de 1000 no pasa nada
    print("---------------------")
```

## ¿Cómo extraer detalles?

```python
# Para extraer todo el html
def get_html(url):
    '''
    Le pregunta el url y se lo retorna en formato texto
    '''
    resp = requests.get(url).text
    return BeautifulSoup(resp, "lxml")
```

```python
def get_date(speech_url):
    '''
    Retorna el año, mes, día de una URL
    '''
    date = re.findall("Paginas/(.{8})", speech_url)[0] 
    # Buscara el número que aparece en la url y tomará el primer elemento
    year = date[0:4] # toma los primeros cuatro digitos
    month = date[4:6] # toma los otros dos 
    day = date[6:8] # Lo últimos 2
    return year, month, day
```

```python
def get_title(soup):
    '''
    Me devuelve el título del speech
    '''
    return soup.title.text.strip()
```

```python
def get_speech(soup):
    '''
    Me devuelve el speech. Generalmente tienen una < p > 
    '''
    return str(soup.find_all('p'))
```

```python
def remove_tags(text):
    '''
    Strips text from html and other tags
    '''
    return re.sub('<[^>]+>|\]|\[|\\u200b|\\n|\xa0', '', text)
```

```python
def get_speech_elements(speech_url):
    '''
    Me retorna los elementos que quiero
    '''
    print('Speech url: ', speech_url) # Me dice que URL estoy utilizando
    soup = get_html(speech_url) # Me obtiene el url
    title = get_title(soup) # El título
    year, month, day = get_date(speech_url) # La fecha
    speech = remove_tags(get_speech(soup)) # El speech
    return [year, month, day, title, speech_url, speech]
```

```python
elements = get_speech_elements(prueba)
elements
```
