```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
# Llamemos la URL que queremos consultar. En este caso es la página de Davivienda.
url = "https://www.davivienda.com/wps/portal/personas/nuevo"
```

Note que es una página compleja. Es decir, tiene bastante movimiento y no es del todo plana.

```python
# Necesitamos este código para acceder al código HTML de la página.
from urllib.request import urlopen

# Capturamos el código HTML
webpage = urlopen(url) 

# Para verlo más lindo
soup = BeautifulSoup(webpage, "html.parser")
print(soup.prettify)
```

Note que es un código extraño. A pesar de que parece HTML es JavaScript

```python
# Recuerde que importamos TIME porque es una forma de hacer que se demore buscando la página.
import time

# Cambiamos la URL
url = 'https://www.davivienda.com/wps/portal/personas/nuevo#iframe'

#Nos permite entrar a códigos JavaScript
# Utilizando Phantom que esta guardado en ese path en MI PC. No necesariamente en otro.
browser1 = webdriver.PhantomJS(executable_path=r"C:\Users\tatoo\Desktop\PhantomJS\phantomjs-2.1.1-windows\bin\phantomjs.exe")
  
browser1.get(url)
html1 = browser1.page_source

time.sleep(5)
```

```python
html1
```

```python
# Hacerlo con Google Chrome
browser2 = webdriver.Chrome(executable_path=r"C:\Users\tatoo\Desktop\PhantomJS\Chrome\chromedriver.exe")
time.sleep(5)
browser2.get(url)
html2 = browser2.page_source
time.sleep(5)
#browser2.quit() # Si utilizamos esto cerramos la pestaña
html2
```

Lo que hace este comando es abrir una pestaña de Google Chrome y meterse a la URL escogida.

```python
# Esta línea permite ejecutar el Javascript, es decir, me lo muestra.
# El orden lógico es empezar por el código de arriba y luego por esto.

html = browser2.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
    
soup = BeautifulSoup(html, "html.parser")
```

```python
soup
```

**Cuando usamos la función execute_script, Python hace el rendering de la página por medio de Phantom**

```python
# Y este si es el código que nos interesa:
print(soup.prettify)
```

**Tipos de Localizadores**
- **ID**	To find the element by ID of the element
- **Classname**	To find the element by Classname of the element
- **Name**	To find the element by name of the element
- **Link text**	To find the element by text of the link
- **XPath**	XPath required for finding the dynamic element and traverse between various elements of the web page

__Fuente: https://www.guru99.com/xpath-selenium.html__

```python
# PASOS -
# 1. Entrar a la página web de NewTours
browser = webdriver.Chrome(executable_path=r"C:\Users\tatoo\Desktop\PhantomJS\Chrome\chromedriver.exe")
browser.get("http://newtours.demoaut.com/")

# 2. Esperar 3 segs
time.sleep(3)

# Note que la página tiene un user y un password. 
# Cuándo uno se mete al código, se da cuenta que User name y Password
# tienen códigos que se llaman name = userName
# Lo que tenemos que hacer es identificar esos espacios
texto_usuario = browser.find_element_by_name("userName")
texto_pass = browser.find_element_by_name("password")
boton_submit = browser.find_element_by_name("login")

# En los lugares que definimos, inserte ese valor.
texto_usuario.send_keys("daniel")
texto_pass.send_keys("prueba")

time.sleep(4)
boton_submit.click()

print("Terminé")
```

```python
browser.quit() # Para que se cierre la pestaña que creamos
print("Prueba completada")
```

**Si queremos que lo haga todo completo y con un orden lógico**

```python
browser = webdriver.Chrome(executable_path=r"C:\Users\tatoo\Desktop\PhantomJS\Chrome\chromedriver.exe") # Sabe que se meterá a Internet
browser.get("http://newtours.demoaut.com/") # URL
browser.maximize_window() # Pone grande la pantalla. El Código JS puede cambiar dependiendo del tamaño
time.sleep(3) # Espera 3 segs depués de meterse
browser.find_element_by_link_text("SUPPORT").click() # Darle click al botón SUPPORT
time.sleep(5)
browser.get("http://newtours.demoaut.com/")
time.sleep(5)
texto_usuario = browser.find_element_by_name("userName")
texto_pass = browser.find_element_by_name("password")
boton_submit = browser.find_element_by_name("login")
texto_usuario.send_keys("daniel")
texto_pass.send_keys("prueba")
boton_submit.click()
time.sleep(4)
browser.quit()
print("Prueba completada")
```
