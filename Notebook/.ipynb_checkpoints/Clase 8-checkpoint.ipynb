```python
# Agrupare por CA a todas las observaciones
crimes_by_community = crimes.groupby("Community Area")
```

No habíamos hablado de lo que se hace después de agregar
- Hay funciones predefinidas, como "sum", "mean", "mode", "count", "std", "unique".
- También se puede aplicar con funciones arbitrarias definidas por el usuario.

```python
community_crime_count = crimes_by_community["ID"].agg("count")
# Coja el df y tome la variable ID, cuantos IDSs hay por Community Area
```

```python
community_crime_count
# Hay 77 comm. area. Note que cada ID es un robo, entonces lo que me está diciendo es que hay tantos delitos por CA
```

```python
community_crime_count.to_frame() # Es mejor verlo siempre así. Se presta para hacer gráficos-
```

```python
community_crime_count.plot(kind='bar'); # Haga un plot de barras con esos datos
```

### Series de tiempo 

```python
crimes.head()
```

```python
# Note que tenemos un formato fecha.
crimes["Day"] = crimes["Date"].dt.day 
# Me crea una columna Day, la cual me va a tomar el día de la columna fecha
```

```python
crimes.Day
```

```python
crimes_by_day = crimes.groupby('Day')
# Agrupa por día
```

```python
crimes_by_day['ID'].agg('count') # cuenta cuantos ID hay por agrupación.
```

```python
crimes_by_day['ID'].agg('count').plot(kind='bar');
```

```python
# Para el CA 41, agrupelos por Día, tome el ID y cuente cuantos hay.
# Después de eso, ploteelo.
crimes[crimes['Community Area'] == 41].groupby('Day')['ID'].agg('count').plot(kind='bar');
```

### Multiple Group by

```python
# Haga un agrupamiento por area y por día. Es decir, vea para cada área, en cada uno de los días el número de robos
crimes_by_community_day = crimes.groupby(['Community Area', 'Day'])
```

```python
crimes_by_community_day_count = crimes_by_community_day['ID'].agg('count')
crimes_by_community_day_count
```

```python
# En vez de contar cuando delitos hay, podemos ver cuantos son capturados.
# sum() sume todos los arrestos que hay.
crimes_by_community_day_arrests = crimes_by_community_day['Arrest'].agg('sum')
crimes_by_community_day_arrests
```

```python
# calculamos el cociente de arrestos sobre delitos
community_day_arrest_prop = crimes_by_community_day_arrests / crimes_by_community_day_count
# Note que laforma de la base es la misma. Solo cambia el último número.
```

## Unstacking

¿Cómo graficar series en una sola pasada?

```python
# Poner algunas como filas. Lo que hace es que cada obs es un día y cada CA una col
community_arrest_timeseries = community_day_arrest_prop.unstack('Community Area')
community_arrest_timeseries
```

```python
community_arrest_timeseries[[40,41,42]].plot(); # Tomamos solos las Community Area 
```
